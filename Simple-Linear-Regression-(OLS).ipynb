```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import time
import warnings
warnings.filterwarnings('ignore')

class linear_reg():
    
    def __init__(self):
        self.m=None
        self.b=None
    
    def fit(self,x_train,y_train):
        
        m_num=0
        m_den=0
        
        for i in range(x_train.shape[0]):   
            m_num+=(x_train[i]-x_train.mean())*(y_train[i]-y_train.mean())
            m_den+=(x_train[i]-x_train.mean())**2
        
        self.m=m_num/m_den
        self.b=(y_train.mean())-(self.m)*(x_train.mean())
        print("\n\n"+'\033[1m'+"THE LINEAR REGRESSION MODEL IS READY"+'\033[0m'+"\n\n")
    
    def predict(self,x_test):
        
        return self.m* x_test + self.b
    
    def coefficient(self):
        
        return self.m
    
    def intercept(self):
        
        return self.b
    
a=input("enter location of data:\n")
print("\n(it is assumed that the first column is the input and the second column is the output of the data)\n\n\n")
df=pd.read_csv(a)

t=1
time.sleep(t)
print(df.head())

time.sleep(t)
print("\n\nthe graph of the plot is: ")
sns.scatterplot(data=df, x=df.columns[0],y=df.columns[1])
plt.show()
print("\n")

x=df.iloc[:,0].values
y=df.iloc[:,1].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)

lr=linear_reg()

time.sleep(t)
lr.fit(x_train,y_train)
time.sleep(t)

def plot():
    sns.scatterplot(data=df,x=df.columns[0],y=df.columns[1])
    plt.plot(x,lr.predict(x),color="red")
    plt.show()
    print("\n")

def pred():
    u=float(input("\ninput: "))
    print("the predicted output is: ",lr.predict(u))
    print("\n")
    
while True:
    q=int(input("type 1 to get prediction\ntype 2 to get the line graph\ntype 3 to get the coefficient\ntype 4 to get the intercept\ntype 5 to get out of here\nyour response: "))
    if q==1:
        pred()
        print("\n")
        
    elif q==2:
        plot()
        print("\n")
        
    elif q==3:
        print("\nthe coeffient is: ",lr.coefficient())
        print("\n")
        
    elif q==4:
        print("\nthe intercept is: ",lr.intercept())
        print("\n")
        
    elif q==5:
        break
        
    else:
        print("not a valid response?\n")
        continue
```
